## Q1.請使用以下資訊建立網絡，並且繪圖

| from | to | weight |
|:---:|:---:|:---:|
| 0 | 5 | 0.5 |
| 1 | 3 | 0.8 |
| 2 | 1 | 0.2 |
| 3 | 4 | 1.0 |
| 4 | 5 | 0.4 |
| 5 | 2 | 0.1 |
| 3 | 0 | 0.3 |

```python
import networkx as nx

g1 = nx.DiGraph()


g1.add_weighted_edges_from([(0, 5, 0.5), (1, 3, 0.8), (2, 1, 0.2), (3, 4, 1.0), (4, 5, 0.4), (5, 2, 0.1), (3, 0, 0.3)])
weight = nx.get_edge_attributes(g1, "weight").values()

nx.draw_networkx(g1, width=list(weight))
```

## Q2.請針對第1題所建立的網絡，呈現它的鄰近矩陣

```python
matrix = nx.adjacency_matrix(g1)
matrix.todense()
```
