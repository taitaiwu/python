```python
import networkx as nx
from operator import itemgetter
```

## Q1.請使用以下資訊建立有方向的網絡，並且計算密度與繪圖

| from | to | weight |
|:---:|:---:|:---:|
| 0 | 5 | 0.5 |
| 1 | 3 | 0.8 |
| 2 | 1 | 0.2 |
| 3 | 4 | 1.0 |
| 4 | 5 | 0.4 |
| 5 | 2 | 0.1 |
| 3 | 0 | 0.3 |

```python
g1 = nx.DiGraph()


g1.add_weighted_edges_from([(0, 5, 0.5), (1, 3, 0.8), (2, 1, 0.2), (3, 4, 1.0), (4, 5, 0.4), (5, 2, 0.1), (3, 0, 0.3)])
weight = nx.get_edge_attributes(g1, "weight").values()

nx.draw_networkx(g1, width=list(weight))
print("Density :", nx.density(g1))
```

## Q2.請針對第1題所建立的網絡，計算5種中心性並將結果排序

```python
print("Degree Centrality :", sorted(nx.degree_centrality(g1).items(),key=itemgetter(1), reverse=True)) 
print("Closeness Centrality :", sorted(nx.closeness_centrality(g1).items(),key=itemgetter(1), reverse=True))
print("Between Centrality :", sorted(nx.betweenness_centrality(g1).items(),key=itemgetter(1), reverse=True))
print("Eigenvector Centrality", sorted(nx.eigenvector_centrality(g1).items(),key=itemgetter(1), reverse=True))
print("Page Rank :", sorted(nx.pagerank(g1).items(), key=itemgetter(1), reverse=True))
```
