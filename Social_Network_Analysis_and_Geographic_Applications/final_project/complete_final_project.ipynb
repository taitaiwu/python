## Event 視角分析

```python
df = pd.read_csv("complete_data.csv")
df.drop("姓名", axis=1, inplace=True)
n = df.values.T@df.values
list1 = list(df)
list2 = [list1[i] for i in range(len(list1))]

print("鄰近矩陣：")
print("\t", "\t".join(list2))

for i in range(len(n)):
    str1 = list2[i] + "\t"
    
    for j in n[i]:
        str1 += str(j) + "\t"

    print(str1)

print("\n=======================================================================\n")

jaccard = [[0 for _ in range(len(n))] for _ in range(len(n))]

for i in range(len(n)):
    for j in range(len(n)):
        jaccard[i][j] = n[i][j] / (n[i][i] + n[j][j] - n[i][j])

print("Jaccard 正規化：")  
print("\t", "\t".join(list2))

for i in range(len(jaccard)):
    str1 = list2[i] + "\t"
    
    for j in jaccard[i]:
        str1 += str(round(j, 2)) + "\t"

    print(str1)
```

### 中心性分析

```python
df2 = pd.read_csv("familiarity_questionnaire.csv", index_col=0)
g1 = nx.from_pandas_adjacency(df2, create_using=nx.DiGraph())

print("Degree Centrality :", sorted(nx.degree_centrality(g1).items(),key=itemgetter(1), reverse=True)) 
print("Closeness Centrality :", sorted(nx.closeness_centrality(g1).items(),key=itemgetter(1), reverse=True))
print("Between Centrality :", sorted(nx.betweenness_centrality(g1).items(),key=itemgetter(1), reverse=True))
print("Eigenvector Centrality", sorted(nx.eigenvector_centrality(g1).items(),key=itemgetter(1), reverse=True))
```

### clique

```python
g2 = nx.Graph()

for i, j in g1.edges():  # 提取出彼此都認識的人
    if (g1.has_edge(i, j)):
        g2.add_edge(i, j)

cliques = [x for x in nx.find_cliques(g2)] 

for i in range(len(cliques)):
    print("clique" + str(i) + ": \t" + ", ".join(cliques[i]))
```

### K-core

```python
g3 = g2.to_undirected()
g3.remove_edges_from(nx.selfloop_edges(g3))  # 移除自我迴圈

counter = 0

while True:
    core = nx.k_core(g3, k=counter)
    counter += 1

    if not core:
        best_k = counter - 2
        best_core = nx.k_core(g3, k=best_k)
        print("K-core :", best_k)
        print("節點數 :", best_core.number_of_nodes())
        print("節點 :", ", ".join(list(best_core.nodes)))
        break

```
