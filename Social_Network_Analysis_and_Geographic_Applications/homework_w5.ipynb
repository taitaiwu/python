## Q1.請針對networkx內建的網絡（florentine families graph）回答下列問題：

```python
import networkx as nx

g = nx.florentine_families_graph()
nx.draw(g, with_labels=True)
```

### 1. 用Louvain method可以找到幾個社群？

```python
print("Number of communities :", len(nx.community.louvain_communities(g)))
```

### 2. 每個社群內部的成員有誰？

```python
for i in range(len(nx.community.louvain_communities(g))):
    print("Community", i+1, ":", nx.community.louvain_communities(g)[i], end="\n")
```

### 3.請用Block model的形式繪製分群後的網絡

```python
partition = []    

for i in nx.community.louvain_communities(g):
    partition.append(list(i))

M = nx.quotient_graph(g, partition, relabel=True) 
nx.draw_networkx(M)
```
